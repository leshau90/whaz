```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('~/jupyter-portofolio/nilai-tukar-sgd-idr-2021-2022.csv')
dailydf = pd.read_csv('~/jupyter-portofolio/nilai-tukar-sgd-idr-2021-2022-daily.csv')
```

```python

plt.plot(df['Date'],df['Close'])
plt.xticks(rotation=45)
miny=min(df['Close'])
for x, y in zip(df['Date'],df['Close']):
    plt.plot([x, x], [miny, y], linestyle='--', color='gray', linewidth=0.8)
# plt.figure(figsize=(10,10))
plt.title("Nilai tukar SGD > IDR per Bulan 2021-2022")
plt.savefig('./images/monthly-graph.png')

plt.show()
```
